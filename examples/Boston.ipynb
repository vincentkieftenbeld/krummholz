```python
from sklearn.datasets import load_boston

import numpy as np
```

```python
from saplings.training import train_test_split
from saplings.criteria import MAE, MSE
from saplings.metrics import mean_squared_error
from saplings.gradient_boosting import GradientBoostedTrees
from saplings.plot import plot_training
```

```python
boston = load_boston()
X = boston.data 
y = boston.target

X_train, X_test, y_train, y_test = train_test_split(X, y)
```

```python
mean_squared_error(y, np.mean(y))/2
```

```python
model = GradientBoostedTrees(criterion=MSE(), n_estimators=40, learning_rate=1, max_depth=2)
model.fit(X_train, y_train)

plot_training(model, X_test, y_test)

y_pred = model.predict(X_test)
MSE().loss(y_test, y_pred)
```

```python
model = GradientBoostedTrees(criterion=MSE(), n_estimators=40, learning_rate=1.0, max_depth=1)
model.fit(X_train, y_train)

plot_training(model, X_test, y_test)

y_pred = model.predict(X_test)
MSE().loss(y_test, y_pred)
```
